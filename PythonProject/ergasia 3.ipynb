```python
import pandas as pd
import itertools

# Define levels and factor names
levels = [-1, 1]
factor_names = ['Factor1', 'Factor2', 'Factor3', 'Factor4']

# Generate full factorial design
design = list(itertools.product(levels, repeat=4))

# Create base DataFrame
df = pd.DataFrame(design, columns=factor_names)

# Add coded levels (0/1 instead of -1/1)
coded_df = df.replace({-1: 0, 1: 1})
coded_df.columns = [f + '_coded' for f in factor_names]

# Add interaction terms (2-way interactions)
df['F1xF2'] = df['Factor1'] * df['Factor2']
df['F1xF3'] = df['Factor1'] * df['Factor3']
df['F1xF4'] = df['Factor1'] * df['Factor4']
df['F2xF3'] = df['Factor2'] * df['Factor3']
df['F2xF4'] = df['Factor2'] * df['Factor4']
df['F3xF4'] = df['Factor3'] * df['Factor4']

# Add 3-way interactions
df['F1xF2xF3'] = df['Factor1'] * df['Factor2'] * df['Factor3']
df['F1xF2xF4'] = df['Factor1'] * df['Factor2'] * df['Factor4']
df['F1xF3xF4'] = df['Factor1'] * df['Factor3'] * df['Factor4']
df['F2xF3xF4'] = df['Factor2'] * df['Factor3'] * df['Factor4']

# Add 4-way interaction
df['F1xF2xF3xF4'] = df['Factor1'] * df['Factor2'] * df['Factor3'] * df['Factor4']

# Add empty Response column
df['Response'] = ''

# Combine original, coded, and interaction columns
full_df = pd.concat([df, coded_df], axis=1)

# Save to Excel
full_df.to_excel('4_factor_2_level_design_full_template.xlsx', index=False)

print("Excel file created: 4_factor_2_level_design_full_template.xlsx")
```
